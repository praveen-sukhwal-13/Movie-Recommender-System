Feature Selection Part ( Feature Reduction)

```python
# now we select only important feature or we do feature reduction
# so important feature to recommend a movies is its id, title, genre, iverview # this are the important feature of the movies
movies = movies[['id', 'title', 'overview', 'genre']]
```

movies

```python
movies # now we have only important feature we discarded the feature which were irrelevant 
```

```python
# we are doing context based perdictions so we will use tags for it 
# for tags we will merge overview and genre together
movies['tags'] = movies['overview']+movies['genre']
```

```python
movies
```

```python
# apn ne overview and genre ko add krke tags bnya hai tw ab apn un dono ko drop krnge q ki yha data replicate ho rha hai
# to do such we do the below things
new_data = movies.drop(columns = ['overview', 'genre']) # movies ke column se overview and genre ko drop krdo
```

```python
new_data
```

now we will convert the text data into some vector and apply some algorithms on it

to convert data into vector we have two methods
1. Bag of word
2. TFIDF
These are the two techniques to convert tags text into vector

```python
from sklearn.feature_extraction.text import CountVectorizer # converting text into vector
```

```python
cv = CountVectorizer(max_features=10000, stop_words='english')# language is english
```

```python
cv
```

```python
# now we have to fit the transform data
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray() # new data ke tags ko univalue m convert krne ke bd array me convert krna hai
```

```python
vector.shape
```

10000 rows and 10000 column

```python
# now suppose we are watching iron man movie and now our system should recommend the movie Avengers also because both 
# the movie has same context
# to find similarity between movies we will use COSINE SIMILARITY
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector) # vector ko similarity me convert kra
```

```python
similarity
```

```python
# ab apn ko us similar movie ka index number chiye q ki jo movie apn ko recommend hogi voh index number ke basis pe hogi
# and based on the index number I can access the movie title
```

```python
new_data[new_data['title'] == 'The Godfather'].index[0]
```

```python
new_data.info()
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse = True, key = lambda vector:vector[1]) # har movies ke index ke bich ki similarity
for i in distance[:5]:
    print(new_data.iloc[i[0]].title) # 5 similar movies # gives the title of the index number # 0 isliye q ki index 0 p hai
```

```python
def recommand(movies):
    index = new_data[new_data['title'] == movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse = True, key = lambda vector:vector[1]) # har movies ke index ke bich ki similarity
    for i in distance[:5]:
        print(new_data.iloc[i[0]].title) # 5 similar movies # gives the title of the index number # 0 isliye q ki index 0 p hai
```

```python
recommand("Iron Man")
```
